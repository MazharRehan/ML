```python
# Task: Predict the per capita income of Canada for the year 2020
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year, df.per_capita_income, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression() # linearRegression object created
reg.fit(df[['year']], df.per_capita_income) # Train the model
```

```python
# Predict the per capita income for the year 2020
reg.predict(pd.DataFrame([[2020]], columns=['year']))
```

```python
# predict the per capita income from 2017 to 2050
years = pd.DataFrame(list(range(2017, 2051)), columns=['year'])
prediction = reg.predict(years)
```

```python
years['per_capita_income'] = prediction
years
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year, df.per_capita_income, color='red', marker='+')
plt.plot(years.year, years.per_capita_income, color='blue')
```
